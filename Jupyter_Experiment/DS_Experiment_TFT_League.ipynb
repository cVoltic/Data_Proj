```python
#DE Libraries
import pandas as pd
import numpy as np
#Function to treat JSON Column
def JSONParser(data):
    import json
    j1=json.loads(data)
    return j1

#Fix Formatting for JSON Data single '' to ""
#df.str.replace("\'", "\"")

#Data Engineering JSON column
#Read in master table
masterGameDF = pd.read_csv("DataStorage/TFT_League/TFT_Master_MatchData.csv",
                           converters={'combination':JSONParser,'champion':JSONParser})

masterGameDF
```

```python
#combination and champion JASON to DF treatment
#convert combination to champion origin group dataFrame
champGroup = pd.DataFrame(list(masterGameDF['combination'][0:]))

#reanme some headers & fill NaN with 0
champGroup.columns = champGroup.columns.str.replace("Set3_", "")

champGroup = champGroup.fillna(0)
champGroup = champGroup.astype(int)
champGroup
```

```python
# setup champion group dataframe
# split group/class
originDF = pd.read_csv("DataStorage/TFT_League/TFT_Champion_CurrentVersion.csv",usecols=['origin'],header=0).drop_duplicates()
originDF = originDF[0:]
originDF['origin']=originDF['origin'].str.replace(" ", "")
originDF['origin']=originDF['origin'].str.replace("-", "")
originList = list(originDF['origin'])

# sort champGroupDF by origin/Classes
originGroupDF = pd.DataFrame(champGroup,columns=originList)
classGroupDF = champGroup.drop(originList, axis=1)

print(len(originList))
# origin DF
print(originGroupDF)
# class DF
print(classGroupDF)
```

```python
#Treat Champion and items combination
from pandas.io.json import json_normalize

#Flatten JASON
champItemStarList = list(masterGameDF['champion'])
champItemStarDF = pd.json_normalize(champItemStarList)
champItemStarDF
```

```python
#Split items/stars df, Rename, Transpose
items = [col for col in champItemStarDF.columns if '.items' in col]
itemsDF = champItemStarDF[items]
itemsDF.columns = itemsDF.columns.str.replace(".items", "")
itemsDF

stars = [col for col in champItemStarDF.columns if '.star' in col]
starsDF = champItemStarDF[stars]
starsDF.columns = starsDF.columns.str.replace(".star", "")
starsDF
```

```python
# treat NaN as no items
for label, content in itemsDF.items():
    '''
        #For each row in DF index
        #where in the current column, there is an NaN value
        #fill it with a empty list
    '''
    
    for row in itemsDF.loc[itemsDF[label].isnull(), str(label)].index:
        itemsDF.at[row, str(label)] = []
    
itemsDF

# for label, content in starsDF.items():
#     '''
#         #For each row in DF index
#         #where in the current column, there is an NaN value
#         #fill it with a empty list
#     '''
    
#     for row in starsDF.loc[starsDF[label].isnull(), str(label)].index:
#         starsDF.at[row, str(label)] = []
    
# starsDF
```
